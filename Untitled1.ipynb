```python
import pandas as pd

# Create the first DataFrame
data1 = {'product': ['A', 'B', 'C', 'A', 'B'],
        'customer': ['X', 'Y', 'Z', 'X', 'W']}
df1 = pd.DataFrame(data1)

# Create the second DataFrame
data2 = {'product': ['A', 'B', 'C', 'D'],
        'price': [10, 20, 30, 40]}
df2 = pd.DataFrame(data2)

# Merge the two DataFrames based on the 'product' column
merged_df = pd.merge(df1, df2, on='product', how='inner')  # Use 'inner' join

merged_df
```

```python
# prompt: inner join

# Merge the two DataFrames based on the 'product' column
merged_df = pd.merge(df1, df2, on='product', how='inner')

# Display the merged DataFrame
merged_df
```

```python
# Merge the two DataFrames based on the 'product' column using outer join
merged_df = pd.merge(df1, df2, on='product', how='outer')

# Display the merged DataFrame
merged_df
```
