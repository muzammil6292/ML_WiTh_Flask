```python
#normalization
#with min max scaler
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

#sample dataset with multiple features
data = {
    'Age': [25, 30, 35, 40],
    'Income': [50000, 60000, 70000, 80000],
    'height': [160, 170, 180, 190],
    'weight': [60, 70, 80, 90]
}

df = pd.DataFrame(data)
print(df)

#initialize minmaxscaler
scaler = MinMaxScaler()

#normalize the dataset
normalized_data = scaler.fit_transform(df)

#create a new dataframe with the normalized values
normalized_df = pd.DataFrame(normalized_data, columns=df.columns)
print("\nNormalize DataFrame (scales to range [0,1]): ")
print(normalized_df)
```

```python
#standardization
from sklearn.preprocessing import StandardScaler
import pandas as pd

#sample dataset with different features
data={
    'Age': [25,30,45,50,28,36],
    'Income': [50000,60000,70000,80000,55000,65000]
}

df = pd.DataFrame(data)
print(df)

#initialize standardization with data
scaler = StandardScaler()

#standardize the dataset
standardized_data = scaler.fit_transform(df)

#create a new dataframe with the standardized values
standardized_df = pd.DataFrame(standardized_data, columns=df.columns)
print("\nStandardized DataFrame: ")
print(standardized_df)
```
