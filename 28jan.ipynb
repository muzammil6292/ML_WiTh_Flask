```python
#label encoding
import pandas as pd
from sklearn.preprocessing import LabelEncoder

#sample data
data={
    "customer_id":[1,2,3,4,5,6,7],
    "gender":['male','male','female','male','female','male','female'],
    "age":[25,30,35,40,45,50,55],
    "subscription_status":["active","inactive","active","inactive","active","inactive","active"]
}
df=pd.DataFrame(data)
df

#initiate label encoder in data
le=LabelEncoder()
df["age_encoded"]=le.fit_transform(df["age"])
df["gender_encoded"]=le.fit_transform(df["gender"])
df

```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Sample data
data = {
    "customer_id": [1, 2, 3, 4, 5, 6, 7],
    "gender": ['male', 'male', 'female', 'male', 'female', 'male', 'female'],
    "age": [25, 30, 35, 40, 45, 50, 55],
    "subscription_status": ["active", "inactive", "active", "inactive", "active", "inactive", "active"]
}

# Create DataFrame
df = pd.DataFrame(data)

# Initiate one-hot encoding
ohe = OneHotEncoder(sparse_output=False)
gender_encoded = ohe.fit_transform(df[["gender"]])
gender_encoded_df = pd.DataFrame(gender_encoded, columns=ohe.get_feature_names_out(["gender"]))
subscription_status_encoded = ohe.fit_transform(df[["subscription_status"]])
subscription_status_encoded_df = pd.DataFrame(subscription_status_encoded, columns=ohe.get_feature_names_out(["subscription_status"]))

# Display encoded DataFrame
gender_encoded_df
subscription_status_encoded_df
```
