```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(1000)
data=np.random.randint(10000,5000000,20000000)
#create a histogram
plt.hist(data, bins=1000, edgecolor="black", color="red")
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
#find the mean pf the data
mean=np.mean(data)
print("Mean:",mean)
#find the median of the data
median=np.median(data)
print("Median:",median)
#find the standard deviation of the data
std=np.std(data)
print("Standard Deviation:",std)
# create a boxplot
plt.boxplot(data)
plt.title("Boxplot")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
