```python
import pandas as pd
import numpy as np

data = pd.DataFrame({
    'Brand': ['Maruti', 'Hyundai', 'Tata', 'Mahindra', 'Maruti', 'Hyundai',
              'Renault', 'Tata', 'Maruti'],
    'Year': [2012, 2014, 2011, 2015, 2012, 2016, 2014, 2018, 2019],
    'Kms Driven': [50000, 30000, 60000, 25000, 10000, 46000, 31000, 15000, 12000],
    'City': ['Gurgaon', np.nan, 'Mumbai', 'Delhi', 'Mumbai', 'Delhi',
             'Mumbai', 'Chennai', 'Ghaziabad'],
    'Mileage': [28, 27, 25, 26, 28, 29, 24, 21, 24]
})

# Display the DataFrame
display(data)

# Filter where Brand is 'Maruti' and Year is 2012
display(data.loc[(data['Brand'] == 'Maruti') & (data['Year'] == 2012)])

# Filter where Brand is 'Hyundai' and 'Kms Driven' > 30000
display(data.loc[(data['Brand'] == 'Hyundai') & (data['Kms Driven'] > 30000)])

# Filter where City is 'Delhi' and 'Kms Driven' > 30000
display(data.loc[(data['City'] == 'Delhi') & (data['Kms Driven'] > 30000)])

# Creating another DataFrame for joins
customers = pd.DataFrame({
    'Brand': ['Maruti', 'Hyundai', 'Tata', 'Mahindra', 'Renault'],
    'Customer Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Discount Amount': [5000, 7000, 4500, 8000, 6000]
})

# Inner Join
result_inner = pd.merge(data, customers, on='Brand', how='inner')
display(result_inner)

# Left Join
result_left = pd.merge(data, customers, on='Brand', how='left')
display(result_left)

# Right Join
result_right = pd.merge(data, customers, on='Brand', how='right')
display(result_right)

# Outer Join
result_outer = pd.merge(data, customers, on='Brand', how='outer')
display(result_outer)
```
